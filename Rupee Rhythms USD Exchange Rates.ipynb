# Report on Historical Analysis of PKR, INR, and USD Exchange Rates (1947-2024)

**Introduction:**
This report provides a comprehensive analysis of the exchange rate dynamics between the Pakistani Rupee (PKR), Indian Rupee (INR), and the United States Dollar (USD) from 1947 to 2024. The dataset used for this analysis includes yearly exchange rate values for PKR and INR against USD, allowing for insights into the economic narratives of Pakistan and India in reference to the USD.

**Data Overview:**
The dataset consists of three columns: Year, INR (Rs), and PKR (Rs), totaling 78 entries. It captures the exchange rates of PKR and INR against USD over the specified timeframe, providing a rich source of information for historical analysis.

**Analysis and Insights:**

1. **Trend Analysis:**
   - A line plot depicting the trend of INR and PKR against USD reveals fluctuating exchange rates over the years. Both currencies show varying degrees of appreciation and depreciation against the USD, influenced by economic, geopolitical, and global factors.
   - Overall, the trend suggests a long-term depreciation of PKR and INR against USD, with periods of volatility and stability.

2. **Relationship Analysis:**
   - A scatter plot illustrating the relationship between INR and PKR against USD showcases their correlation. While both currencies generally move in the same direction against USD, there are instances where one currency outperforms the other.
   - The scatter plot helps identify periods of divergence or convergence between PKR and INR exchange rates.

3. **Decadal Analysis:**
   - A bar plot presenting the average exchange rate of INR and PKR against USD for every decade provides insights into long-term trends. It highlights the average performance of each currency over distinct periods, reflecting economic developments and policy changes.
   - Decadal analysis allows for a macro-level understanding of currency performance and stability.

4. **Distribution Analysis:**
   - A box plot visualizing the distribution of exchange rates for INR and PKR against USD showcases the variability and spread of exchange rate data. It reveals the presence of outliers, central tendencies, and the overall distribution shape for each currency.
   - Understanding the distribution of exchange rates aids in assessing the level of volatility and risk associated with currency fluctuations.

5. **Frequency Distribution Analysis:**
   - A histogram displaying the frequency distribution of exchange rates for INR and PKR against USD provides insights into the frequency of occurrence of different exchange rate levels. It helps identify common exchange rate ranges and assess the degree of stability or volatility in currency values.
   - Frequency distribution analysis aids in understanding the probability of specific exchange rate movements and potential outcomes.

**Conclusion:**
The historical analysis of PKR, INR, and USD exchange rates from 1947 to 2024 offers valuable insights into the economic trajectories of Pakistan and India in relation to the global economy. By examining trends, relationships, and distributions in exchange rate data, researchers, economists, policymakers, investors, and businesses can gain a deeper understanding of the factors shaping currency dynamics and make informed decisions in their respective domains.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/kaggle/input/exchange-rate-pkr-inr-usd-1947-2024/rupee_vs_dollar.csv")
```

```python
# Visualization 1: Line Plot showing the trend of INR and PKR against USD over the years
plt.figure(figsize=(10, 6))
plt.plot(data['year'], data['ind(Rs)'], label='INR (Rs)')
plt.plot(data['year'], data['pk(Rs)'], label='PKR (Rs)')
plt.xlabel('Year')
plt.ylabel('Exchange Rate against USD')
plt.title('Exchange Rate Trend (1947-2024)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Visualization 2: Scatter Plot showing the relationship between INR and PKR against USD in a particular year
plt.figure(figsize=(8, 6))
plt.scatter(data['ind(Rs)'], data['pk(Rs)'], alpha=0.7)
plt.xlabel('INR (Rs) against USD')
plt.ylabel('PKR (Rs) against USD')
plt.title('INR vs PKR against USD (1947-2024)')
plt.grid(True)
plt.show()
```

```python
# Visualization 3: Bar Plot showing the average exchange rate of INR and PKR against USD for every decade
data['decade'] = (data['year'] // 10) * 10
avg_rates = data.groupby('decade').mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(avg_rates['decade'], avg_rates['ind(Rs)'], width=5, label='INR (Rs)')
plt.bar(avg_rates['decade'], avg_rates['pk(Rs)'], width=5, label='PKR (Rs)', alpha=0.7)
plt.xlabel('Decade')
plt.ylabel('Average Exchange Rate against USD')
plt.title('Average Exchange Rate per Decade (1947-2024)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Visualization 4: Box Plot showing the distribution of exchange rates for INR and PKR against USD
plt.figure(figsize=(8, 6))
plt.boxplot([data['ind(Rs)'], data['pk(Rs)']], labels=['INR (Rs)', 'PKR (Rs)'])
plt.ylabel('Exchange Rate against USD')
plt.title('Distribution of Exchange Rates (1947-2024)')
plt.grid(True)
plt.show()
```

```python
# Visualization 5: Histogram showing the frequency distribution of exchange rates for INR and PKR against USD
plt.figure(figsize=(8, 6))
plt.hist(data['ind(Rs)'], bins=20, alpha=0.7, label='INR (Rs)')
plt.hist(data['pk(Rs)'], bins=20, alpha=0.7, label='PKR (Rs)')
plt.xlabel('Exchange Rate against USD')
plt.ylabel('Frequency')
plt.title('Exchange Rate Frequency Distribution (1947-2024)')
plt.legend()
plt.grid(True)
plt.show()
```
